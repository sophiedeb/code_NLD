```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
a=0.14
b=0.6
```

```python
def glycolysis(x, t):
    nx0 = -x[0]+a*x[1]+x[0]**2*x[1]
    nx1 = b-a*x[1]-x[0]**2*x[1]
    res = np.array([nx0, nx1])
    return res
```

```python
a
```

```python
ts = np.linspace(0.0, 50.0, 5000)

xs = odeint(glycolysis, [0.2, 3], ts)
plt.plot(xs[:,0], xs[:,1])
xs = odeint(glycolysis, [0, 0.1], ts)
plt.plot(xs[:,0], xs[:,1])
xs = odeint(glycolysis, [3.0, 3.0], ts)
plt.plot(xs[:,0], xs[:,1])
plt.gca().set_aspect('equal')
plt.savefig('vanderpol_oscillator.png')
plt.show() 
```
